```python
def linear_regression(data, target, corrl):
    
    # y = b1 * x + b0
    
    data = data.astype(float)
    
    x = data[target].tolist()
    
    b0 = - 0.5
    
    if bool(data[target].corr(data[corrl]) > 0) is True:
        
        b1 = 1 / median(x)
        
    else: 
        
        b1 = -1 / median(x)
    
    true_positives = 0
    false_positives = 0
    true_negatives = 0
    false_negatives = 0
    
    for value in x:
        
        y = b1 * value + b0
        
        if bool(y >= 0.5 and value >= median(x)) is True:
            true_positives += 1
        elif bool(y <= 0.5 and value >= median(x)) is True:
            false_negatives += 1
        elif bool(y <= 0.5 and value <= median(x)) is True:
            true_negatives += 1
        else:
            false_negatives += 1
            
    return [true_positives, false_positives, true_negatives, false_negatives]
```

```python
bismillah = linear_regression(ctcf, "LLR", "tfbs")
bismillah
```

```python
def logistic_regression(data, target, corrl):
    
    # y = b1 * x + b0
    
    data = data.astype(float)
    
    x = data[target].tolist()
    
    b0 = - 0.5
    
    if bool(data[target].corr(data[corrl]) > 0) is True:
        
        b1 = 1 / median(x)
        
    else: 
        
        b1 = -1 / median(x)
    
    true_positives = 0
    false_positives = 0
    true_negatives = 0
    false_negatives = 0
    
    for value in x:
        
        y = 1 / 1 + math.exp(-(b1 * value + b0))
        
        if bool(y >= 0.5 and value >= 0.5 * median(x)) is True:
            true_positives += 1
        elif bool(y <= 0.5 and value >= 0.5 * median(x)) is True:
            false_negatives += 1
        elif bool(y <= 0.5 and value <= 0.5 * median(x)) is True:
            true_negatives += 1
        else:
            false_negatives += 1
            
    return [true_positives, false_positives, true_negatives, false_negatives]
```

```python
logistic_bismillah = logistic_regression(ctcf, "LLR", "tfbs")
logistic_bismillah
```
